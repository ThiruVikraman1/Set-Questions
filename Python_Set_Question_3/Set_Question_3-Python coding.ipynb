```python
 #3.Find Even and Odd Numbers in a List
Question: Write a Python program to separate even and odd numbers from the list [12, 17, 19, 24, 35, 40, 56].

```

```python
list1 = [12, 17, 19, 24, 35, 40, 56]
even = []
odd = []
for num in list1:
    if num%2 == 0:
        even.append(num)
    else:
        odd.append(num)
print("even numbers: ",even)
print("odd numbers: ",odd)
```

```python
#4. Reverse a Number
Question: Write a Python program to reverse the number 12345.

```

```python
num = 12345
num1 = str(num)
rev = num1[::-1]
print(rev)
```

```python
#5. Find the Second Largest Number in a List
Question: Write a Python program to find the second largest number in the list [12, 45, 98, 34, 23].
```

```python
list1 =  [12, 45, 98, 34, 23]
list1.sort()
sec_large = list1[-2]
print(sec_large)
```

```python
# different method
list1 =  [12, 45, 98, 34, 23]
large_number = max(list1)
sec_large = list1[0]
for i in list1:
    if i != large_number and i > sec_large:
        sec_large = i
print(sec_large)
```
